```python
import requests
from bs4 import BeautifulSoup
import time

# 결과를 담을 리스트
all_books = []

# 1페이지부터 3페이지까지 반복
for page in range(1, 4):
    # f-string을 이용해 URL의 페이지 번호
    url = f"https://www.yes24.com/Product/Category/BestSeller?categoryNumber=001&pageNumber={page}"
    
    res = requests.get(url)
    soup = BeautifulSoup(res.text, "html.parser")
    
    # 책 제목들 찾기
    titles = soup.select("a.gd_name")
    
    for t in titles:
        all_books.append(t.get_text().strip())
    
    print(f"{page}페이지 수집 완료!")
    
    time.sleep(1)

print("-" * 30)
print(f"총 수집된 책 개수: {len(all_books)}권")
print(all_books[:10]) # 상위 10개만 확인
```

```python
import requests
from bs4 import BeautifulSoup

# 1. 네이버 뉴스 (IT/과학) 주소
url = "https://news.naver.com/section/105" 

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

# 2. 페이지 가져오기
res = requests.get(url, headers=headers)
soup = BeautifulSoup(res.text, "html.parser")

# 3. 뉴스 제목 요소 찾기
titles = soup.select('strong.sa_text_strong')

# 4. 태그를 뺀 제목만 저장할 빈 리스트 만들기
clean_titles = []

for title in titles:
    # .get_text()로 태그 안의 글자만 가져오기
    clean_titles.append(title.get_text().strip())

print(f"저장된 제목 개수: {len(clean_titles)}개")
print(clean_titles[:10])
```
