### 2. 서울시 거주/비거주 고객 비교 분석

```python
len(sample_df[sample_df.CSTMR_MEGA_CTY_NM == '서울특별시']) # 서울시 거주 고객 수
len(sample_df[sample_df.CSTMR_MEGA_CTY_NM != '서울특별시']) # 서울시 비거주 고개 수
sample_df[sample_df.CSTMR_MEGA_CTY_NM == '서울특별시']['AMT'].sum() # 서울시 거주 고객 총 소비액
# sample_df.loc[sample_df.CSTMR_MEGA_CTY_NM == '서울특별시', 'AMT'].sum()
sample_df[sample_df.CSTMR_MEGA_CTY_NM != '서울특별시']['AMT'].sum() # 서울시 비거주 고객 총 소비액
sample_df[sample_df.CSTMR_MEGA_CTY_NM == '서울특별시']['CNT'].sum() # 서울시 거주 고객 카드 이용 건수
sample_df[sample_df.CSTMR_MEGA_CTY_NM != '서울특별시']['CNT'].sum() # 서울시 비거주 고객 카드 이용 건수
```

```python
seoul_dict = {'서울시' : ['거주', '비거주'],
              '고객 수' : [54150, 45851],
              '총 소비액' : [119663142676, 146587135822],
              '카드 이용 건수' : [5542462, 4950200]}
seoul_df = pd.DataFrame(seoul_dict)
seoul_df
```

### 3. 편의점 소비 정보 분석

```python
sample_df[sample_df.TP_BUZ_NO == 4010]['AMT'].sum() # 전지역 편의점 소비액
```

```python
sample_df[sample_df.MEGA_CTY_NM == '서울특별시']['AMT'].sum() # 가맹점 주소 기준 서울시 소비액
```

```python
sample_df[(sample_df.TP_BUZ_NO == 4010) & (sample_df.CTY_RGN_NM == '강남구')]['AMT'].sum() # 강남구 편의점 소비액
```

```python
sample_df[(sample_df.TP_BUZ_NO == 4010) & (sample_df.CTY_RGN_NM == '강남구') & (sample_df.CSTMR_CTY_RGN_NM == '강남구')]['AMT'].sum() # 강남구 거주자의 강남구 편의점 소비액
```

```python
sample_df[(sample_df.TP_BUZ_NO == 4010) & (sample_df.CTY_RGN_NM == '강남구') & (sample_df.CSTMR_CTY_RGN_NM != '강남구')]['AMT'].sum() # 강남구 비거주자의 강남구 편의점 소비액
```

```python
seoul_con_dict = {'강남구 편의점' : ['거주자', '비거주자'],
                  '소비액' : [288379890, 418895250]}
seoul_con_df = pd.DataFrame(seoul_con_dict)
seoul_con_df
```

```python
seoul_con_dict = {'강남구 편의점 총 소비액' : [707275140],
                  '거주자' : [288379890],
                  '비거주자' : [418895250]}
seoul_con_df = pd.DataFrame(seoul_con_dict, index = ['소비액'])
seoul_con_df
```
